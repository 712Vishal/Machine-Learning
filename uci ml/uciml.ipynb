# normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y)
```

# K NEAREST NEIGHBOUR

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
knn_pred[:10]
```

```python
# pd.get_dummies(y['Class'],prefix='Class',dtype=int)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test, knn_pred)*100
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, knn_pred)
cm
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y_pred = log_model.predict(x_test)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)*100
```

```python
log_model.score(x_train, y_train)*100
```

# support vector machine

```python
from sklearn.svm import SVC

svm=SVC(kernel='linear')
svm.fit(x_train, y_train)
```

```python
svm_pred = svm.predict(x_test)
svm_pred[:10]
```

```python
accuracy_score(y_test, svm_pred)*100
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt='d', )
```

# Naive bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(x_train, y_train)
```

```python
naive_pred = nb_clf.predict(x_test)
naive_pred[:10]
```

```python
accuracy_score(y_test, svm_pred)*100
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(confusion_matrix(y_test, naive_pred), annot=True, fmt='d', )
plt.show()
```
